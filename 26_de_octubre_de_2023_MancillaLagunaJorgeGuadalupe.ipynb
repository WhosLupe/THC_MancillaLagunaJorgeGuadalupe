```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2 * np.pi, 100)

y_seno = np.sin(x)
y_coseno = np.cos(x)

plt.plot(x, y_seno, color='k', lw = '2')
plt.plot(x, y_coseno, color='c')

plt.xlabel('x')
plt.ylabel('y')

plt.title('Seno y Coseno')

plt.grid(ls = ':')
plt.show()
```

```python
def graficacion(x):
  x = np.linspace(0, 2 * np.pi, 100)

  y_seno = np.sin(x)
  y_coseno = np.cos(x)

  plt.plot(x, y_seno, color='k', lw = '2')
  plt.plot(x, y_coseno, color='c')

  plt.xlabel('x')
  plt.ylabel('y')

  plt.title('Seno y Coseno')

  plt.grid(ls = ':')
  plt.show()
graficacion(y)
```
