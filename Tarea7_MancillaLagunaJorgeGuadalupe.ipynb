1.- Con base en la definición de número primo, escriba una función que reciba un número entero e imprima 'True' si es primo o de lo contrario imprima 'False'. Tenga en cuenta que si el número es menor o igual a 1, se retorna como resultado "Error en el Número" y debe volver a preguntar hasta que la entrada sea valida.

```python
def es_primo(num):
    if num <= 1:
        return "Error en el Número"
    elif num == 2:
        return True
    elif num % 2 == 0:
        return False
    else:
        for i in range(3, int(num**0.5) + 1, 2):
            if num % i == 0:
                return False
        return True

#Se pide al usuario que ingrese un número hasta que sea válido
while True:
    try:
        num = int(input("Ingrese un número entero mayor que 1: "))
        sol = es_primo(num)
        print(sol)
        break  #Sale del bucle si el número es válido
    except error:
        print("Error: Ingrese un número entero válido.")
```

2.- Sin utilizar librerías adicionales cree un arreglo de numéros enteros (entre 0 y 150) y calcule a través de funciones estándar o lambda lo siguiente:

*   El promedio
*   La varianza
*   La desviacion estandar
*   Visualice los datos en un histograma.
*   La suma de los números pares e impares.
*   Los múltiplos de 2, 3 y 5.










```python
#Arreglo de numeros
num = list(range(151))

#Promedio
prom = sum(num) / sum(1 for _ in num)

#Varianza
vari = sum((x - prom) ** 2 for x in num) / sum(1 for _ in num)

#Desviación estándar
DE = vari ** 0.5

#Histograma
print("Histograma de Números entre 0 y 150:")
for num in range(151):
    cont = num.count(num)
    if cont > 0:
        print(f'{num}: {"*" * num}')

#Suma de los números pares e impares
SP = sum(x for x in num if x % 2 == 0)
SI = sum(x for x in num if x % 2 != 0)

#Múltiplos de 2, 3 y 5
multi2 = [x for x in num if x % 2 == 0]
multi3 = [x for x in num if x % 3 == 0]
multi5 = [x for x in num if x % 5 == 0]

#Resultados
print(f'\nPromedio: {prom}')
print(f'Varianza: {vari}')
print(f'Desviación Estándar: {DE}')
print(f'Suma de Números Pares: {SP}')
print(f'Suma de Números Impares: {SI}')
print(f'Múltiplos de 2: {multi2}')
print(f'Múltiplos de 3: {multi3}')
print(f'Múltiplos de 5: {multi5}')
```

3.- En la clase de ecuaciones diferenciales se pide a los estudiantes encontrar la solución del problema de valor inicial:

\begin{equation}
y´=-y+x \hspace{3cm} y(-1)=5.
\end{equation}


Los estudiantes llegaron a la solución general cuya ecuación es:

\begin{equation}
y=(x-1)+C e^{-x}
\end{equation}

Apoye a sus compañeros y grafique en python la familia de curvas para la solución de la ecuación diferencial y en un sólo gráfico muestre de diferente color las curvas para los siguientes tres casos:

*   para valores de C > 0
*   cuando C = 0
*   para valores de C < 0

Señale en el gráfico las curvas que corresponden a cada uno de los tres casos.

```python
import numpy as np
import matplotlib.pyplot as plt

#Función que representa la solución general
def SG(x, C):
    return (x - 1) + C * np.exp(-x)

#Valores de x para la gráfica
x_vals = np.array([i/100 for i in range(-300, 301)])  # equivalente a np.linspace(-3, 3, 600)

#C > 0
plt.plot(x_vals, SG(x_vals, 2), label='C > 0', color='blue')

#C = 0
plt.plot(x_vals, SG(x_vals, 0), label='C = 0', color='green')

#C < 0
plt.plot(x_vals, SG(x_vals, -2), label='C < 0', color='red')

#Grafica
plt.title('Familia de Curvas para la Ecuación Diferencial')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

4.- Sea la función:
\begin{equation}
\frac{3x + 7}{x^2 - 4}
\end{equation}

Utilizando la librería matplotlib:

1.- Grafique la función con línea continua para un rango de -10 a 10, con más de 1000 muestras.

2.- Investigue en la literatura cómo graficar las asíntotas horizontales y verticales. Muéstrelas en el gráfico con líneas punteadas.

3.- Establezca límites en el eje "y" para evitar que la gráfica muestre valores muy altos o muy bajos cerca de las asíntotas verticales.

4.- Añada una leyenda indicando los elementos presentes en la gráfica.

```python
import numpy as np
import matplotlib.pyplot as plt

#Funcion
def F(x):
    return (3*x + 7) / (x**2 - 4)

#Valores de x
x_vals = np.linspace(-10, 10, 1000)

#Se grafica la función
plt.plot(x_vals, F(x_vals), label='Función', color='blue')

#Asíntotas verticales
plt.axvline(x=-2, linestyle='--', color='red', label='Asíntota Vertical x = -2')
plt.axvline(x=2, linestyle='--', color='red', label='Asíntota Vertical x = 2')

#Asíntota horizontal
plt.axhline(y=0, linestyle='--', color='green', label='Asíntota Horizontal y = 0')

#Límites en el eje y
plt.ylim(-10, 10)

#Leyenda
plt.legend()

#Creacion gráfica
plt.title('Gráfica de la Función con Asíntotas')
plt.xlabel('x')
plt.ylabel('y')

# Mostrar la gráfica
plt.show()
```
